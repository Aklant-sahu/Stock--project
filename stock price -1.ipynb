```python
def clean(data):
    col=['OPEN ','HIGH ','LOW ','PREV. CLOSE ','ltp ','close ','vwap ','52W H ','52W L ']
    data['Date ']=pd.to_datetime(data['Date '])
    data=data.sort_values(by='Date ')
    data.set_index(pd.DatetimeIndex(data['Date ']), inplace=True)
    try:
        data=data.drop(['Date ','series ','VALUE '],axis=1)
    except:
        data=data.drop(['Date '],axis=1)
    for i in col:
        for j in range(len(data)):

            try:
                x=float(data[i][j])
            except:
                x=float(data[i][j].replace(',',''))
            data[i][j]=x
    data=data.astype(float)
    
    return data
```

```python
def emarsi(data):
    data.ta.rsi(close='close ', length=10, append=True)
    data.ta.ema(close='close ', length=10, append=True)
    data=data[10:]
    return data
```

```python
def datashift(data):
    
    col=['OPEN ','HIGH ','LOW ','ltp ','vwap ','VOLUME ','No of trades ','RSI_10','EMA_10']
    ### ltp== last traded price
    ### vwap= volume weighted average price
    data=data.reset_index(drop=True)
    for j in col:
        x=[]
        for i in range(len(data)):
            if (i==0):
                x.append(0)
            else:
                x.append(data[j][i-1])
        nam='prev_'+j
        data[nam]=x
    
    
    data=data.drop(col,axis=1)
    data=data[1:]
    data=data.reset_index(drop=True)
    #data=data.drop(['PREV. CLOSE '],axis=1)
    return data
```

```python
def xy(data):
    X=data.drop('close ',axis=1)
    y=data['close ']
    return X,y
```

```python
def buysell(Xg,yg,pred):
    data=pd.DataFrame(Xg)
    data['current_close']=yg
    data['pred']=pred
    data.reset_index(inplace=True)
    act=[]
    
    for i in range(len(data)):
        
        if (float(data['current_close'][i]) > data['PREV. CLOSE '][i]):
            act.append(1)
        else:
            act.append(0)

    data['action']=act
    act=[]
    for i in range(len(data)):
        if (data['pred'][i]>data['PREV. CLOSE '][i]):
            act.append(1)
        else:
            act.append(0)

    data['action_pred']=act
    return data

    
    
```
